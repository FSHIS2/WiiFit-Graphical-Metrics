```python
df = pd.read_excel (r'/content/drive/My Drive/Colab Notebooks/data'
                      r'/WiiFit GRAPHICAL METRICS PROJECT/SEPTIEMBRE 2021'
                        r'/MARCHA.xlsx')
```

```python
if len(df.columns) == 2:
  display()
else:
  aux_display()
```

```python
def display():
  x_label = str(df.columns[0])
  y_label = str(df.columns[1])

  x_values = list(df.loc[:, x_label])
  y_values = list(df.loc[:, y_label])

  plt.plot(x_values, y_values, 'ro')
  plt.plot(x_values, y_values)
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.grid()
  plt.show()
```

```python
def aux_display():
  date_label = str(df.columns[0])
  left_label = str(df.columns[1])
  right_label = str(df.columns[2])
  balance_label = str(df.columns[3])

  date_values = list(df.loc[:, date_label])
  left_values = list(df.loc[:, left_label])
  right_values = list(df.loc[:, right_label])
  balance_values = list(df.loc[:, balance_label])

  plt.plot(date_values, left_values, color = 'green', label = 'izquierda')
  plt.plot(date_values, right_values, color = 'red', label = 'derecha')
  plt.plot(date_values, balance_values, color = 'blue', label = 'equilibrio')

  plt.xlabel(date_label)
  plt.ylabel('Puntuación')
  plt.legend()
  plt.grid()
  plt.show()
```
